**Import Libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d
import cv2
```

**Load and Displaying the Original Image:**

```python
# Load the image
image = cv2.imread('/content/Teaser Pedigree Cats.jpg', cv2.IMREAD_GRAYSCALE)
# Display the original image-
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.show()
```

**Displaying the Convolution Kernel**

```python
# Define a 3x3 edge detection kernel
kernel = np.array([[1, 0, -1],
             2          [1, 0, -1],
                       [1, 0, -1]])
kernel
```

** Apply convolution**

```python
convolved_image = cv2.filter2D(image, -1, kernel)
convolved_image
```

 **Visualize the results**

```python
plt.subplot(1, 3, 2)
plt.imshow(kernel, cmap='gray')
plt.title('Convolution Kernel')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(convolved_image, cmap='gray')
plt.title('Convolved Image')
plt.axis('off')
plt.show()
```

```python
# Print kernel and image shapes for clarity
print(f"Kernel shape: {kernel.shape}")
print(f"Original image shape: {image.shape}")
print(f"Convolved image shape: {convolved_image.shape}")
```
